```python
import pandas as pd
```

```python
from meteostat import Stations, Daily
```

```python
from datetime import datetime
```

```python
import folium
```

```python
import geopandas as gpd
```

```python
import matplotlib.pyplot as plt
```

```python
start = datetime(2020, 1, 1)
```

```python
end = datetime(2020, 12, 31)
```

```python
stations = Stations()
```

```python
station1 = stations.nearby(51.1657, 10.4515).fetch(1)
```

```python
station2 = stations.nearby(48.2082, 16.3738).fetch(1)
```

```python
data1 = Daily(station1.index[0], start, end).fetch()
```

```python
data2 = Daily(station2.index[0], start, end).fetch()
```

```python
avg_temp1 = data1['tavg'].mean()
```

```python
avg_temp2 = data2['tavg'].mean()
```

```python
print(f'Average temperature in Germany: {avg_temp1}째C')
```

```python
print(f'Average temperature in Austria: {avg_temp2}째C')
```

```python
data1['tavg'].plot(label='Germany') 

data2['tavg'].plot(label='Austria') 

plt.legend() 

plt.show() 
```

```python
m = folium.Map(location=[50, 10], zoom_start=5)
```

```python
folium.Marker([51.1657, 10.4515], popup='Germany').add_to(m)
```

```python
folium.Marker([48.2082, 16.3738], popup='Austria').add_to(m)
```
